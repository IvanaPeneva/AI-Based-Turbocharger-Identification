```python
from google.colab import drive
drive.mount('/content/drive')

import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow

input_dir = '/content/drive/MyDrive/allRotatedPhotos'
output_dir = '/content/drive/MyDrive/medianAllRotatedPhotos'


os.makedirs(output_dir, exist_ok=True)
valid_images = [f for f in os.listdir(input_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

kernel_size = 7

print(f"Found {len(valid_images)} images. Applying {kernel_size}x{kernel_size} median filter and displaying all...")
print("="*50)

for filename in valid_images:
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename)

    image = cv2.imread(input_path)

    if image is not None:
        processed_image = cv2.medianBlur(image, kernel_size)

        cv2.imwrite(output_path, processed_image)

    else:
        print(f"Warning: Could not read file: {filename}")

print("\nProcessing complete!")
```
