```python
import os
import json
import cv2
import easyocr

image_folder = '/content/drive/MyDrive/allRotatedPhotos'
output_json = '/content/drive/MyDrive/analys_masks.json'

print("Initializing EasyOCR reader...")
reader = easyocr.Reader(['en'])

extracted_data = []

if not os.path.exists(image_folder):
    print(f"Error: The folder '{image_folder}' does not exist. Please check your Google Drive mount.")
else:
    valid_extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.tiff')

    for filename in os.listdir(image_folder):
        if filename.lower().endswith(valid_extensions):
            image_path = os.path.join(image_folder, filename)
            image = cv2.imread(image_path)

            if image is None:
                print(f"Warning: Could not read {filename}. Skipping.")
                continue

            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            results = reader.readtext(gray)

            extracted_text = " ".join([text for (bbox, text, prob) in results])

            extracted_data.append({
                "filename": filename,
                "extracted_text": extracted_text
            })

            print(f"Processed: {filename} | Extracted {len(results)} text blocks")
    with open(output_json, 'w', encoding='utf-8') as json_file:
        json.dump(extracted_data, json_file, indent=4, ensure_ascii=False)

    print("-" * 45)
    print(f"Processing complete! Data successfully saved to:\n{output_json}")
```
