```python
import os
import json
import pytesseract
from PIL import Image

image_folder = '/content/drive/MyDrive/allRotatedPhotos'
output_json = '/content/drive/MyDrive/analys_masks.json'

ocr_results = []

valid_extensions = ('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')

print(f"Scanning directory: {image_folder}")

if os.path.exists(image_folder):
    for filename in os.listdir(image_folder):
        if filename.lower().endswith(valid_extensions):
            image_path = os.path.join(image_folder, filename)

            try:
                img = Image.open(image_path)
                extracted_text = pytesseract.image_to_string(img)
                ocr_results.append({
                    "filename": filename,
                    "extracted_text": extracted_text.strip()
                })
                print(f"Successfully processed: {filename}")

            except Exception as e:
                print(f"Error processing {filename}: {e}")

    os.makedirs(os.path.dirname(output_json), exist_ok=True)

    with open(output_json, 'w', encoding='utf-8') as f:
        json.dump(ocr_results, f, indent=4, ensure_ascii=False)

    print(f"\nFinished! Results saved to: {output_json}")

else:
    print(f"Error: The directory {image_folder} was not found. Please check the path and ensure your Google Drive is mounted.")
```
